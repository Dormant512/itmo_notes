### Descriptive Statistics

Find the sample mean $\overline X$:

```python
data['Shots on Goal 1'].mean()
```

Find the sample median $Me$:

```python
data['Shots on Goal 1'].median()
```

Find the variance estimate $S^2$:

```python
data['Shots on Goal 1'].var(ddof = 0)
```

Find the standard deviation  $\sigma$:

```python
data['Shots on Goal 1'].std(ddof = 0)
```
