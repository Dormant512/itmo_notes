```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df_candy = pd.read_csv('./candy-data.csv')

predictors = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']

df_candy_test = pd.read_csv('./candy-test.csv')

df_candy = df_candy[df_candy['competitorname'] != 'Boston Baked Beans']
df_candy = df_candy[df_candy['competitorname'] != 'Dum Dums']
df_candy = df_candy[df_candy['competitorname'] != 'Fruit Chews']

print(f"Length = {len(df_candy)}")

df_candy.head()
```

```python
print(len(df_candy_test))
df_candy_test.head()
```

```python
pd.concat([df_candy_test[df_candy_test['competitorname'] == 'Tootsie Roll Midgies'], \
          df_candy_test[df_candy_test['competitorname'] == 'Trolli Sour Bites']]).head()
```

```python
X = np.zeros((len(df_candy), len(predictors)))
for num, p in enumerate(predictors):
    X[:,num] = df_candy[p]
    
X_test = np.zeros((len(df_candy_test), len(predictors)))
for num, p in enumerate(predictors):
    X_test[:,num] = df_candy_test[p]
    
Y = np.array(df_candy['Y']).reshape(-1,1)

candy_model = LogisticRegression(random_state=2019, solver='lbfgs').fit(X, Y)

Y_pred = list(candy_model.predict(X_test))
Y_real = list(df_candy_test['Y'])

Y_roc = list(candy_model.predict_proba(X_test)[:,1])

print(Y_pred)
print(Y_real)
print(candy_model.predict_proba(X_test)[8])
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(Y_real, Y_pred).ravel()
precision = tp / (tp + fp)
recall = tp / (tp + fn)
print(precision, recall)
```

```python
print(f"ROC AUC SCORE {metrics.roc_auc_score(Y_real, Y_roc)}")

fpr, tpr, thresh = metrics.roc_curve(Y_real, Y_roc)
ans = metrics.auc(fpr, tpr)

print(f"AUC standalone {ans}")
```
