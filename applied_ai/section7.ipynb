```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df_shop = pd.read_csv('./lec7_line_time.csv')
df_shop.head()
```

```python
print(f"X.mean = {df_shop['X'].mean()}")
print(f"Y.mean = {df_shop['Y'].mean()}")
```

```python
X, Y = np.array(list(df_shop['X'])).reshape(-1,1), np.array(list(df_shop['Y'])).reshape(-1,1)
shop_model = LinearRegression().fit(X, Y)
```

```python
print(f"Theta 1 = {shop_model.coef_}")
print(f"Theta 0 = {shop_model.intercept_}")
print(f"R^2 = {shop_model.score(X, Y)}")
```

```python
df_candy = pd.read_csv('./candy-data.csv')

predictors = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']

df_candy_test = pd.concat([df_candy[df_candy['competitorname'] == 'One dime'], \
                          df_candy[df_candy['competitorname'] == 'Nestle Butterfinger']])

df_candy = df_candy[df_candy['competitorname'] != 'One dime']
df_candy = df_candy[df_candy['competitorname'] != 'Nestle Butterfinger']

print(f"Length = {len(df_candy)}")

df_candy.head()
```

```python
df_candy_test.head()
```

```python
X = np.zeros((len(df_candy), len(predictors)))
for num, p in enumerate(predictors):
    X[:,num] = df_candy[p]
    
X_test = np.zeros((len(df_candy_test), len(predictors)))
for num, p in enumerate(predictors):
    X_test[:,num] = df_candy_test[p]
    
Y = np.array(df_candy['winpercent']).reshape(-1,1)

candy_model = LinearRegression().fit(X, Y)
print(candy_model.predict(X_test))
last = np.array([[0, 1, 1, 1, 1, 1, 1, 0, 1, 0.26, 0.296]])
print(candy_model.predict(last))
```
